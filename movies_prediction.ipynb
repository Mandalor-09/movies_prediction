```python
import ast
def converting_genres_into_list(genres_row_data):
  empty_list =[]
  genres_row_data = ast.literal_eval(genres_row_data)
  for every_dictionary in genres_row_data:
    empty_list.append(every_dictionary['name'])
  print(empty_list)
  return empty_list
```

```python
import ast

def cast_keywords_into_list(cast_row_data):
    empty_list = []
    cast_row_data = ast.literal_eval(cast_row_data)
    i = 0
    for every_dictionary in cast_row_data:
        if i < 3:
            empty_list.append(every_dictionary['name'])
            i += 1
        else:
            break
    return empty_list
```

```python
import ast

def crew_into_list(crew_row_data):
    director = []
    crew_row_data = ast.literal_eval(crew_row_data)
    for every_dictionary in crew_row_data:
      if every_dictionary['job']=='Director':
        director.append(every_dictionary['name'])

    return director

```

# **Steaming**

```python
from nltk.stem.porter import PorterStemmer
import nltk
from nltk.corpus import stopwords
import re

nltk.download('stopwords')
words = stopwords.words('english')

portstemm = PorterStemmer()
def stemming(content):
  # if used this re.sub('[a-zA-Z]',' ',content) than we will get all numbers ,puntuation mark like this output will be . . , : ( ) - . . , , “ ” 2018. , . . , ’ “ ,” , . , , , . , - “ ” , , , . “ ( ) : ‘ ’ - ... 7 ,’” , , . “ , ’ , 10 11 . , ’ . ... ’ ,” . ’ - , $1.5 10 $20 . “ ’ ,” . . . $1.5 , , . “ ... ’ . ’ , ,” . , . . 30 , . , , , “ ,” , 2018. , “ ” , , , , . ’ , . , . , , - , “ ,” . 2018 , , , . . , . “ ,” . , . . 6 7, . , . $81 , , . $44 . . :
  removing_numbers = re.sub('[^a-zA-Z]',' ',content)
  lowering =removing_numbers.lower()
  spliting = lowering.split()
  stemmed_value = [portstemm.stem(word) for word in spliting if not word in words]
  stemmed_final_value = ' '.join(stemmed_value)
  return stemmed_final_value



```

```python
dataset['tags'] = dataset['tags'].apply(stemming)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
feature_extraction = CountVectorizer(max_features=5000)
vecterizer = feature_extraction.fit_transform(dataset['tags'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vecterizer)
similarity.shape
```

```python
def recommendation(movie):
    movie_index = dataset[dataset['title'] == str(movie)].index[0]
    distance = similarity[movie_index]
    movie_list = sorted(list(enumerate(distance)), reverse=True, key=lambda x: x[1])[1:6]
    for movie in movie_list:
        print(dataset.iloc[movie[0]]['title'])

recommendation('batman')
```
